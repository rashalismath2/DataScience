<b>Python is a high-level, interpreted, interactive and object-oriented scripting language. Okaaay, thats a lots of information right there. Lets breaks these down.</b>

Python is highlevel- Python easier to read, write, and maintain. To run the program thats written in a high level language, we have to translate that code in to machine language by a compiler or an interpreter (Interpreter explained in next point). 

Python is Interpreted − Python is processed at runtime by the interpreter. You do not need to compile your program before executing it. This is similar to PERL and PHP.

Python is Interactive − You can actually sit at a Python prompt and interact with the interpreter directly to write your programs.

Python is Object-Oriented − Python supports Object-Oriented style or technique of programming that encapsulates code within objects. Object oriented programming are mid lievel programming concepts. We will talk about OOP at the last in this series before we begin to learn about pythin libraries. 

Python is a Beginner's Language − Python is a great language for the beginner-level programmers. But not to underestimate pythong is heavily used in so many area in computer science to write programs.
